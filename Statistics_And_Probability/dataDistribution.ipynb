## Probability Density Functions

### Uniform Distribution

The <b>Uniform Distributions</b> is a probability distribution in which every value between an interval from a to b is equally likely to occur. And values outside of the range will never occure.<br>
Characteristics:
- It has fixed number of outcomes.
- All the outcomes are equally likely to occur.

P(x) = 1/n

where:<br>
x = discrete value<br>
n = number of values in the range

Some examples are as below:
- Guessing a Birthday of random people in a street
- Rolling a Die
- Deck of Cards
- Spinning a Spinner
- Lottery Drawings
- Resource Allocation
- Random Number Generation

Suppose we want to study the behaviour of a specific system. The result should be randomly varied between the range -10 to 10 for each simulation run. We run the simulation 100,000 times to gather sufficient data for our study.

```python
%matplotlib inline

stop = -10
start = 10
number_of_runs = 100000
values = np.random.uniform(stop, start, number_of_runs)
plt.hist(values, bins=50)
plt.show()
```

Just to check all the values are within the specified range.

```python
np.all((values <= 10.0) & (values >=-10))
```

### Normal / Gaussian Distribution

The <b>Normal Distribution</b> is the most commonly-used probability distribution in all of statistics. It represents random variables that their distribution is unknown.<br>

Empirical rule of Normal distribution is that, the 68.27% values lies within one standard deviation ,the 95.45% values lies within two standard deviation and the 99.73% values lies within two standard deviation.<br>

Some of its characteristics are:
- Bell shaped
- Symmetrical
- Unimodal – it has one “peak”
- Mean and median are equal; both are located at the center of the distribution
- About 68% of data falls within one standard deviation of the mean
- About 95% of data falls within two standard deviations of the mean
- About 99.7% of data falls within three standard deviations of the mean

Some examples of Normal Distribution are:
- Height and Weights
- Test Scores
- Blood Pressure
- Natural Phenomena such as the distribution of the behavior of particles
- Financial markets such as stock returns
- IQ scores

```python
from scipy.stats import norm
```

In the below example Standard Deviation and Mean values are default which are 1 and 0 respectively.

```python
x = np.arange(-3, 3, 0.0001)
plt.plot(x, norm.pdf(x))
```

### Exponential PDF/ "Power Law" Distribution

The <b>Exponential Distribution</b> is a probability distribution that is used to model the time we must wait until 
a certain event occurs.
If a random variable X follows an Exponential Distribution, then the cumulative density function of X can be written as:

F(x; λ) = 1 – e-λx

where:<br>
λ: the rate parameter (calculated as λ = 1/μ)<br>
e: A constant roughly equal to 2.718

Some examples are as below:
- Time Between Customers enter a certain shop
- Natural Phenomena for example time between earthquakes
- Time Between Customer Calls at different business
- Failure Times of a system

```python
from scipy.stats import expon

x = np.arange(0, 10, 0.0001)
```

Suppose we are analyzing the time it takes for customers to arrive at a store or the inter-arrival time between events in a system, and we suspect that the distribution follows an exponential pattern. To investigate this, we can plot the probability density function (PDF) of the exponential distribution within the range of 0 to 10.

```python
plt.plot(x, expon.pdf(x))
plt.show()
```

The plot provides a clear representation of the distribution's exponential decay, with higher probabilities for smaller values and a decreasing likelihood as the values increase.

## Probability Mass Functions

This functions deal with discrete data.

### Binomial Probability

The <b>Binomial Distribution</b> is a probability distribution that is used to model the probability that a certain number of “successes” occur during a certain number of trials.
The name is Binomial because there are exactly two outcomes of a trial; <I>Success</I> or <I>Failure</I>.

Some examples of where we use BP:
- Number of Side Effects from Medications
- Number of Fraudulent Transactions
- Number of Spam Emails per Day
- Shopping Returns per Week
- Flip Coin
- Quality Control
- Election Outcomes
- Sports Statistics

```python
from scipy.stats import binom

x = np.arange(0, 10, 0.0001)
n, p = 10, 0.5
```

p is the probability of the specific event happens (success), n is number of trials, and x is the desired observation.

In the below example, given the possibility of success is 0.5,
what are the odds that a value from x range will be observed if we repeat the trial 10 times? 

```python
binomProb = binom.pmf(x, n, p)
print(binomProb, "\n")
plt.plot(x, binomProb)
```

### Poisson Probability Distribution

The <b>Poisson Distribution</b> is a probability distribution that is used to model the probability that a certain number of events occur during a fixed time interval when the events are known to occur independently and with a constant mean rate.

Some examples are:
- Telephone Call Arrivals in a given period at a call center
- Traffic Accidents
- Website Visits
- Defects in Manufacturing
- Number of Network Failures per Week

```python
from scipy.stats import poisson
```

My website gets on average 500 visits per day. What is the odds of getting 550 visits per day?

```python
mu = 500
x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(x, mu))
plt.show()
```

Based on the above graph the probabilty of visitors to be 550 is less than 0.2%

## Below are a few common visualization functions in matplotlib library. It is a quick overview of when, where and how to use each of them.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

### plot() Function

This function creates line plots. Some of the use cases for this function are:
- Visualizing Data Trends, example price over time
- Comparing Multiple Datasets
- Visualizing a Mathematical Function
- Exploring a New DataSet (for EDA)
- Time Series Analysis

```python
x = np.arange(0, 10, 0.0001)
y = norm.pdf(x)
plt.plot(x, y)
plt.show()
```

Comparing 2 datasets

```python
start=-3
stop=10
step=0.1
x = np.arange(start, stop, step)
y1 = norm.pdf(x)
plt.plot(x, y1)

location =2
scale=1
y2 = norm.pdf(x, location, scale)
plt.plot(x, y2) 

plt.show()
```

### pie() Function

This function creates a Pie chart. Pie charts are a popular choice for visualizing categorical data. Although they have limitations and can be misinterpreted if not used appropriately. It's important to consider the number of categories, the clarity of labels, and the readability of small slices when deciding to use a pie chart.<br>

Some use cases for the function are:
- Data Distribution
- Market Share Analysis, percentage breakdown
- Survey Results
- Resource Allocation
- Market Segmentation
- Project Progress
- Budget Allocation
- Demographic Analysis, example: different age groups in a population

The below is an example of Market Segmentation.<br>
Parameter <I>explode</I> is used to highlight a specific segment.<br>
Parameter <I>autopct</I> formats the percentage value to be 1 decimal point.

```python
segments = ['Segment A', 'Segment B', 'Segment C', 'Segment D']
proportions = [0.25, 0.3, 0.2, 0.25]
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']
explode = [0, 0.1, 0, 0]

plt.pie(proportions, labels=segments, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True)

plt.title('Market Segmentation')

plt.show()
```

### bar() Function

This function is used to create bar charts. Bar charts are commonly used to compare categorical or discrete data.<br>
Some usecases for bar charts are as below:
- Comparison of Categories
- Ranking or Sorting, show the top selling products
- Frequency Distribution, occurrence of specific events
- Time-based Data, for example monthly sales
- Comparison of Subcategories, by creating a stacked bar chart
- Visualizing KPIs, such as revenue growth or customer satisfaction scores
- Product Comparison, features, rating and other specifications
- Educational Performance, grade distribution
- Customer Feedback Analysis
- Market Share Analysis, breakdown of different companies or products

```python
#Ranking labels
labels = ['First', 'Second', 'Third', 'Fourth', 'Fifth']

#Ranking scores
scores = [95, 88, 76, 82, 90]
color = ['r', 'g', 'b', 'c', 'm']

plt.bar(labels, scores, color = colors)

plt.xlabel('Ranking')
plt.ylabel('Scores')
plt.title('Ranking Scores')

plt.show()
```

### scatter() Function

scatter() function is used to create scatter plots. It is commonly used to visualize the relationship between two or more continuous variables.<br>
Some common scenarios when we use sctter plot are as below:
- Correlation Analysis
- Data Clustering
- Outlier Detection
- Trend Analysis
- Data Distribution, visualize how the data is spread by concentration of data points
- Multivariate Analysis, relationship between multiple values

In the below example we want to visualize the outlier in our data.

```python
# Generate 100 random datapoints
x = np.random.randn(100)
y = np.random.randn(100)

# Add outlier
x = np.append(x, 15)
y = np.append(y, 8)

# Create the scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Outlier Detection')

# Display the plot
plt.show()
```

Create a scatter plot that represents the age vs time spend watching tv.

```python
np.random.seed(42)
num_samples = 100

# Generate random age values between 18 years and 65 years old
age = np.random.randint(18, 65, num_samples)

# Generate random values for hours of watching TV
time_watching_tv = np.random.normal(2.5, 1.5, num_samples)
# Watching TV hours is is less than or equal 5 hours
time_watching_tv = np.clip(time_watching_tv, 0, 5)

# Create the scatter plot
plt.scatter(age, time_watching_tv)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Time Spent Watching TV (hours)')
plt.title('Age vs. Time Spent Watching TV')

# Display the plot
plt.show()
```

### boxplot() or Whisker Function

This plot is used visualize the distribution and statistical summary of a continuous variable or multiple variables.<br>
They are useful for comparing the distributions of different groups or categories.<br>
By looking at the median and its relevant to the auartiles and whiskers it is easy to see the skewness of data.<br>
Some common usecases for boxplot plots are:
- Data Comparison
- Outlier Detection, datapoints that fall outside the whiskers
- Skewness and Symmetry
- Comparison of Multiple Groups
- Statistical Summary
- Data Distribution Analysis
- Time Series Analysis, yearly distribution of temperature
- Comparing Before and After, customer satisfaction ratings before and after a product redesign
- Data Validation

The below example tend to show how we detect outliers by looking at the boxplot

```python
# Generate random data and some outliers
uniformSkewed = np.random.rand(100) * 100 -40
hight_outliers = np.random.rand(10) * 50 + 100
low_outliers = np.random.rand(10) * -50 - 100

# Concat the data
ata = np.concatenate((uniformSkewed, hight_outliers, low_outliers))

# Create boxplot
plt.boxplot(data)

plt.xlabel('Data')
plt.ylabel('Value')
plt.title('Outlier Detection')

# Display the plot
plt.show()
```

The box shows the 2 inner quartiles. 50% of the data resides in the box. The red line in the box shows the median.<br>
outside of the box within the dotted lines or whiskers there are 25% of the data in each side.<br>
The lines at the end of each whisker are outer quartiles. Anything outside of them are considered as outliers.
